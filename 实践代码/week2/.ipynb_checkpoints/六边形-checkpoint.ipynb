```python
"""六边形网格"""
```

```python
#创建常量
X_REPEAT = 19 #水平镶嵌的次数
Y_REPEAT = 12 # 垂直镶嵌的次数

for y in range(Y_REPEAT):
    #显示六边形的上半部分
    for x in range(X_REPEAT):
        print(r'/ \_', end='')
    print()

    #显示六边形的下半部分
    for x in range(X_REPEAT):
        print(r'\_/ ', end='')
    print()
```

```python
#创建常量
X_REPEAT = 16 #水平镶嵌的次数
Y_REPEAT = 6 # 垂直镶嵌的次数

for y in range(Y_REPEAT):
    #显示六边形的上半部分
    for x in range(X_REPEAT):
        print(r' /  \   ', end='')
    print()
    for x in range(X_REPEAT):
        print(r'/    \__', end='')
    print()

    #显示六边形的下半部分
    for x in range(X_REPEAT):
        print(r'\    /  ', end='')
    print()
    for x in range(X_REPEAT):
        print(r' \__/   ', end='')
    print()
```
