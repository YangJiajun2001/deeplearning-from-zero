```python
import math
import random
import statistics
import matplotlib.pyplot as plt
```

# 利用 math 库计算半径为 2 圆的周长和面积

```python
pi = math.pi
r = 2

print('半径为2圆的周长为：', 2 * pi * r)
print('半径为2圆的面积为：', pi * (r **2))
```

# 利用 math 库绘制 f(x) = exp(x) 函数图像，x 的取值范围为 [−3, 3]

```python
x_start , x_end = -3, 3
num = 100
step = (x_end - x_start)/(num-1)
```

```python
x_array = [x_start + i * step for i in range(num)]
y_array = [math.exp(x_idx) for x_idx in x_array]
```

```python
plt.plot(x_array, y_array, marker = '.',
        markersize = 0,
        markerfacecolor = 'w',
        markeredgecolor = 'k')


plt.show()
```
