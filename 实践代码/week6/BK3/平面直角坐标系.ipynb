```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [4, 0, -5, -4, 6, 0]# 定义x坐标的列表
y = [2, -2, 7, -6, -5, 0]# 定义y坐标的列表
labels = ['A', 'B', 'C', 'D', 'E', 'O']# 定义每个点的标签列表
```

```python
fig, ax = plt.subplots() # 创建图形和坐标轴对象

## 绘制点
plt.plot(x, y, 'x')# 使用“x”标记在坐标轴上绘制点
## 添加标签
for label, i, j in zip(labels, x, y):
    plt.text(i, j + 0.5, label + '({}, {})'.format(i, j)) # 在每个点上方添加标签和坐标

## 设置坐标轴标签
plt.xlabel('x')# 设置x轴标签
plt.ylabel('y')# 设置y轴标签

## 绘制x轴和y轴的水平和垂直线
plt.axhline(y=0, color='k', linestyle='-')# 在y=0处绘制水平线
plt.axvline(x=0, color='k', linestyle='-')# 在x=0处绘制垂直线

## 设置刻度
plt.xticks(np.arange(-8, 8+1, step = 1))# 设置x轴的刻度范围为-8到8，步长为1
plt.yticks(np.arange(-8, 8+1, step = 1))# 设置y轴的刻度范围为-8到8，步长为1

## 设置坐标轴比例
plt.axis('scaled')# 设置坐标轴比例，使x轴和y轴的比例一致

## 设置坐标轴范围
ax.set_xlim(-8, 8)# 设置x轴的显示范围为-8到8
ax.set_ylim(-8, 8)# 设置y轴的显示范围为-8到8

## 隐藏坐标轴边框
ax.spines['top'].set_visible(False)# 隐藏顶部边框
ax.spines['bottom'].set_visible(False)# 隐藏底部边框
ax.spines['left'].set_visible(False)# 隐藏左侧边框
ax.spines['right'].set_visible(False)# 隐藏右侧边框

## 添加网格
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5]) # 添加虚线网格，设置网格线样式和颜色
```
